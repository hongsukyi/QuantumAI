```python
# !pip install pennylane koreanize_matplotlib scikit-learn -q
import pennylane as qml
from pennylane import numpy as np
import numpy as std_np
import matplotlib.pyplot as plt
import koreanize_matplotlib
from IPython.display import clear_output
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
plt.rcParams['axes.unicode_minus'] = False

# ============================================================
# ===== 수정 영역 =====
FEAT_IDX  = [2, 3]   # 사용할 feature 인덱스: [0,1] / [2,3] / [0,2] / [1,3]
N_LAYERS  = 2        # Ansatz 레이어 수: 1 / 2 / 3
STEPS     = 80       # 학습 스텝 수: 30 / 50 / 80 / 100
LR        = 0.05     # 학습률: 0.01 / 0.05 / 0.08
# ===== 수정 끝 =====
# ============================================================

N_QUBITS = 3   # feature 2개 + 교차항 1개

print("=" * 60)
print(" Iris 분류  (AngleEmbedding, 2 feature + 교차항 → 3 qubit)")
print("=" * 60)
print(f"  feature idx: {FEAT_IDX}  |  N_QUBITS={N_QUBITS}  |  N_LAYERS={N_LAYERS}")
print(f"  매핑: [x1, x2, x1*x2]  → AngleEmbedding(RY)")

# ===========================================================
#  1. 데이터 준비
# ===========================================================
iris       = load_iris()
feat_names = [iris.feature_names[i] for i in FEAT_IDX]
X_raw      = iris.data[:, FEAT_IDX]
y          = iris.target

X_scaled = MinMaxScaler((0, std_np.pi)).fit_transform(X_raw)

# 교차항 추가: [x1, x2] -> [x1, x2, x1*x2]
X_aug = std_np.column_stack([
    X_scaled,
    X_scaled[:, 0] * X_scaled[:, 1]
])

X_train, X_test, y_train, y_test = train_test_split(
    X_aug, y, test_size=0.3, stratify=y, random_state=42
)
X_train = np.array(X_train)
X_test  = np.array(X_test)

print(f"\n  feature 이름: {feat_names} + 교차항(x1*x2)")
print(f"  클래스: {list(iris.target_names)}")
print(f"  train: {len(X_train)}개, test: {len(X_test)}개")
print(f"  입력 형상: {X_train.shape}  (샘플 x 3)")

# ===========================================================
#  2. 양자 회로 — AngleEmbedding + SEL
# ===========================================================
dev = qml.device("default.qubit", wires=N_QUBITS)

@qml.qnode(dev)
def circuit(x, w):
    # AngleEmbedding: 각 큐빗에 RY(xi)만 적용 (상호작용 없음)
    # → 교차항 x1*x2는 큐빗2의 RY에서 유일하게 인코딩됨
    qml.AngleEmbedding(x, wires=range(N_QUBITS), rotation='Y')
    # Ansatz: SEL이 얽힘과 표현력 담당
    qml.StronglyEntanglingLayers(w, wires=range(N_QUBITS))
    return [qml.expval(qml.PauliZ(i)) for i in range(N_QUBITS)]

# 회로도 시각화
w_shape  = qml.StronglyEntanglingLayers.shape(N_LAYERS, N_QUBITS)
w_sample = np.array(std_np.random.normal(0, 0.1, w_shape))
fig_circ, _ = qml.draw_mpl(circuit, decimals=2)(X_train[0], w_sample)
fig_circ.suptitle(
    f" AngleEmb(RY) + SEL 회로도  (qubits={N_QUBITS}, layers={N_LAYERS})",
    fontsize=12
)
fig_circ.savefig("./lab9_a.png", dpi=300, bbox_inches="tight")
plt.show()

# ===========================================================
#  3. 학습 함수 정의
# ===========================================================
def one_hot(y_arr, n_class=3):
    oh = std_np.zeros((len(y_arr), n_class))
    for i, c in enumerate(y_arr):
        oh[i, c] = 1.0
    return oh

def cost(w):
    total = np.array(0.0)
    for x, label in zip(X_train, y_train):
        out    = np.array(circuit(x, w))
        target = np.array(one_hot([label])[0])
        total  = total + np.mean((out - (2 * target - 1)) ** 2)
    return total / len(X_train)

def predict(X, w):
    scores = std_np.array([std_np.array(circuit(x, w)) for x in X])
    return std_np.argmax(scores, axis=1)

# ===========================================================
#  4. 학습 루프
# ===========================================================
std_np.random.seed(42)
w = np.array(std_np.random.normal(0, 0.1, w_shape), requires_grad=True)
n_params = int(std_np.prod(w_shape))

print(f"\n  파라미터 형상: {w_shape}  →  총 {n_params}개")
print(f"  인코딩: AngleEmbedding(RY) — 큐빗 간 상호작용 없음")
print(f"  → 교차항 큐빗이 유일한 비선형 정보원\n")

opt     = qml.AdamOptimizer(LR)
history = []

for step in range(STEPS):
    w, loss = opt.step_and_cost(cost, w)
    history.append(float(loss))

    if step % 10 == 0 or step == STEPS - 1:
        tr_acc = float(std_np.mean(predict(X_train, w) == y_train))
        te_acc = float(std_np.mean(predict(X_test,  w) == y_test))
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(7, 4))
        ax.plot(history, 'b-', lw=2)
        ax.set_xlabel("Step")
        ax.set_ylabel("MSE Loss")
        ax.set_title(
            f" 훈련 중...  step {step+1}/{STEPS}  "
            f"train={tr_acc:.3f}  test={te_acc:.3f}"
        )
        ax.set_xlim(0, STEPS)
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# ===========================================================
#  5. 최종 결과
# ===========================================================
train_acc = float(std_np.mean(predict(X_train, w) == y_train))
test_acc  = float(std_np.mean(predict(X_test,  w) == y_test))

print("\n" + "=" * 60)
print("최종 결과")
print("=" * 60)
print(f"  train Acc: {train_acc:.4f}")
print(f"  test  Acc: {test_acc:.4f}")
print(f"  파라미터:  {n_params}개  (형상: {w_shape})")

# --- 학습 곡선 저장 ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(history, lw=2)
ax.set_xlabel("Step")
ax.set_ylabel("MSE Loss")
ax.set_title(f" 학습 곡선 (train={train_acc:.3f}, test={test_acc:.3f})")
ax.grid(True, alpha=0.3)
plt.tight_layout()
fig.savefig("./lab9_b.png", dpi=300, bbox_inches="tight")
plt.show()

# ===========================================================
#  6. 결정경계 시각화
# ===========================================================
margin = 0.2
x_min = X_scaled[:, 0].min() - margin
x_max = X_scaled[:, 0].max() + margin
y_min = X_scaled[:, 1].min() - margin
y_max = X_scaled[:, 1].max() + margin

res = 40
xx, yy = std_np.meshgrid(
    std_np.linspace(x_min, x_max, res),
    std_np.linspace(y_min, y_max, res)
)

# [x1, x2, x1*x2] 형태로 grid 구성
grid_x1 = xx.ravel()
grid_x2 = yy.ravel()
grid_pts = std_np.column_stack([grid_x1, grid_x2, grid_x1 * grid_x2])

Z = predict(np.array(grid_pts), w).reshape(xx.shape)

fig, ax = plt.subplots(figsize=(7, 5))
ax.contourf(xx, yy, Z, levels=[-0.5, 0.5, 1.5, 2.5],
            alpha=0.3, cmap='RdYlBu')
ax.contour(xx, yy, Z, levels=[0.5, 1.5], colors='k', linewidths=0.8)

colors = ['steelblue', 'tomato', 'green']
for c, (name, color) in enumerate(zip(iris.target_names, colors)):
    mask_tr = y_train == c
    mask_te = y_test  == c
    ax.scatter(X_train[mask_tr, 0], X_train[mask_tr, 1],
               c=color, s=30, alpha=0.7,
               label=f'{name} train')
    ax.scatter(X_test[mask_te, 0], X_test[mask_te, 1],
               c=color, s=80, marker='*', edgecolors='black',
               label=f'{name} test')

ax.set_xlabel(feat_names[0])
ax.set_ylabel(feat_names[1])
ax.set_title(
    f" 결정 경계 (AngleEmb + SEL)\n"
    f"train={train_acc:.3f}, test={test_acc:.3f}"
)
ax.legend(fontsize=7, loc='best')
ax.grid(True, alpha=0.3)
plt.tight_layout()
fig.savefig("./lab9_c.png", dpi=300, bbox_inches="tight")
plt.show()
```
